```python
# CALCULATE LIMIT POSITIONS
import matplotlib.pyplot as plt
import numpy as np
import multiprocessing

class NQueens:
    def __init__(self, number_pieces=None, number_spaces=None, row=None, column=None, diagonal=None):
        self.n = number_pieces
        self.s = number_spaces
        self.i = row
        self.j = column
        self.k = diagonal
        
        # Restrictions of spaces
        self.restriction_corners = self.s - 4
        self.spaces_occupied = self.s - self.n

        self.restriction_row = self.i - self.n
        self.restriction_column = self.j - self.n
        self.restriction_diagonal = self.k - self.n

        self.total_space_row = self.s - (self.i - self.n)
        self.total_space_column = self.s - (self.j - self.n)
        self.total_space_diagonal = self.s - (self.k - self.n)

        # Initialize an empty board
        self.board = np.zeros((self.s, self.s), dtype=int)

    def calculate_limit_position(self):
        # Calculate the limit position based on your formula
        limit_position = self.restriction_corners + self.spaces_occupied + self.total_space_row + self.total_space_column + self.total_space_diagonal
        
        return limit_position

# Parameters
board_size = 8
n_queens = 8
    
# Pirnt results
n_queens_instance = NQueens(number_pieces=8, number_spaces=8, row=2, column=3, diagonal=5)
limit_position = n_queens_instance.calculate_limit_position()

limit_position
```
